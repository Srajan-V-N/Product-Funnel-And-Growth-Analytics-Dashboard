```python
import pandas as pd
import mysql.connector
import numpy as np

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="product_analytics"
)

cursor = conn.cursor()

def load_csv(table, path):
    df = pd.read_csv(path)

    # Convert NaN to None (VERY IMPORTANT)
    df = df.replace({np.nan: None})

    cols = ",".join(df.columns)
    placeholders = ",".join(["%s"] * len(df.columns))
    sql = f"INSERT INTO {table} ({cols}) VALUES ({placeholders})"

    cursor.executemany(sql, df.values.tolist())
    conn.commit()
    print(f"{table} loaded successfully")

load_csv("customers", "../data/customers.csv")
load_csv("sessions", "../data/sessions.csv")
load_csv("products", "../data/products.csv")
load_csv("events", "../data/events.csv")
load_csv("orders", "../data/orders.csv")
load_csv("order_items", "../data/order_items.csv")
load_csv("reviews", "../data/reviews.csv")
```
